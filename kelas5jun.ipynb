```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-poster')
%matplotlib inline

f = lambda t, s: np.exp(-t) #ode
h = 0.1 # step size
t = np.arange(0, 1 + h, h)
s0 = -1


s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i],  s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='approximate')
plt.plot(t, -np.exp(-t), 'g', label='exact')
plt.title('approximate and exact solution \ for simple ode')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
# Step size
h = 0.1

# Grid waktu
x = np.arange(0, 0.1 + h, h)  # dari 0 ke 0.1
y0 = 1  # initial condition

# Fungsi turunan dari ODE: y' = x^2 - 4y
def f(x, y):
    return x**2 - 4*y

# Metode Euler Eksplisit
y = np.zeros(len(x))
y[0] = y0

for i in range(0, len(x) - 1):
    y[i + 1] = y[i] + h * f(x[i], y[i])

# Cetak hasil
print("Perkiraan y(0.1) =", y[1])

# Plot grafik (opsional)
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b-', label='Euler Approximation')
plt.title('Euler Method for y\' + 4y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

```python
h = 0.05
x = np.arange(0, 1 + h, h)
y0 = 1

# Fungsi turunan dari ODE
def f(x, y):
    return x**2 - 4*y

# Solusi eksak analitik
def y_exact(x):
    return (x**2)/4 - x/8 + 1/32 + (31/32)*np.exp(-4*x)

# Metode Euler
y = np.zeros(len(x))
y[0] = y0
for i in range(len(x) - 1):
    y[i + 1] = y[i] + h * f(x[i], y[i])

# Plot hasil
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'bo--', label='Approximate')  # Euler
plt.plot(x, y_exact(x), 'g-', label='Exact')  # Solusi eksak
plt.title('Approximate and Exact Solution for $y\' + 4y = x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
